### Ejercicio 1:
- Gráfica una recta de $(-7, -7)$ a $(7, 7)$ de color rosa.

```python
X = np.linspace (-7, 7)
Y = np.cos (X)

plt.plot (X, Y, color = "pink")
plt.grid()
plt.show();
```

### Ejercicio 9:
- Muestra en una gráfica de sectores el número de los que no sobrevivieron por clase (_**Pclass**_).

<div align="center">
   
</div>

```python
# Gráfica de tartas
# df[df["Survived"] ==0].value_counts("Pclass", sort=False) Este es para el conteo
df[df["Survived"] ==0].value_counts("Pclass", sort=False, normalize=True) #Este es para la proporción
```

```python
# normalize=True nos da las proporciones en vez de los conteos
values = df[df["Survived"] ==0].value_counts("Pclass", sort=False, normalize=True)
# labels = ["Clase 1", "Clase 2", "Clase 3"] 
labels = sorted([f"clase {i}" for i in df["Pclass"].unique()]) #Se puede hacer también de esta forma

plt.pie(x=values, 
        labels= ["Clase 1", "Clase 2", "Clase 3"], 
        shadow=True, colors=["green", "red", "cyan"], 
        explode = (0.1, 0.1, 0.1), 
        autopct= "%1.2f%%",
        textprops={"fontsize": 15},
        labeldistance=1.45
        );

#el punto y como se pone para que no salga texto sobre la imagen
```

### Ejercicio 10:
- Muestra en 2 graficas de sectores con _**plt.subplots()**_ la siguiente información:
    1. En la primera gráfica mostrar el número de pasajeros que murieron de cada clase.
    2. En la segunda gráfica mostrar el número de pasajeros que sobrevivieron de cada clase.

<div align="center">
   
</div>

```python
labels = sorted([f"clase {i}" for i in df["Pclass"].unique()])

plt.subplot(1, 2, 1)
values = df[df["Survived"] ==0].value_counts("Pclass", sort=False)
plt.pie(x=values, 
        labels= ["Clase 1", "Clase 2", "Clase 3"], 
        shadow=True, colors=["green", "red", "cyan"], 
        explode = (0.1, 0.1, 0.1), 
        autopct= "%1.2f%%")

plt.subplot(1, 2, 2)
values = df[df["Survived"] ==0].value_counts("Pclass", sort=True)
plt.pie(x=values, 
        labels= ["Clase 1", "Clase 2", "Clase 3"], 
        shadow=True, colors=["green", "red", "cyan"], 
        explode = (0.1, 0.1, 0.1), 
        autopct= "%1.2f%%");

plt.title("Not survived vs. Survived %"); #Para ponerle título a la gráfica
```
